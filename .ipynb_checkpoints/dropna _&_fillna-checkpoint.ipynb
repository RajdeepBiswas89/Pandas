**DROPNA AND FILLNA FUNCTIONS**

```python
import pandas as pd
var1=pd.read_csv("D:\\Pandas(python library)\\demo.csv")
var1
```

```python
var1.dropna(axis=0) #deleting the missing data in row order
```

```python
var1.dropna(axis=1) #deleting the missing data in column order
```

thresh functions= if thres=2 then it will drop the elements having a row which has 2 null values

```python
var1.dropna(thresh=2)
```

```python
var1.fillna("python")
```

```python
var1.fillna(
    {
        "a":"rajdeep",
        "b":"rohan",
        "c":"xmas",
        "d":"dhoni"
    }
)
```
